```python
boss_health_images = [f for f in os.listdir(output_folder) if f.endswith(".png")]

fig, ax = plt.subplots(len(boss_health_images), 1, figsize=(20, 10))
for image in boss_health_images:
    img = cv2.imread(output_folder + image)
    width, height = img.shape[1], img.shape[0]
    y_start = int(height // 3)
    height = int(height * 0.4)
    y_end = y_start + height
    x_start = int(width // 3)
    width = int(width * 0.4)
    x_end = x_start + width
    img = img[y_start:y_end,x_start:x_end]


    ax[boss_health_images.index(image)].imshow(img)
    ax[boss_health_images.index(image)].set_title(boss_health_images.index(image))

fig.tight_layout()
plt.show()
```
