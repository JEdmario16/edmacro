```python
boss_health_images = os.listdir(output_folder)

fig, ax = plt.subplots(len(boss_health_images), 1, figsize=(20, 10))
for image in boss_health_images:
    img = cv2.imread(output_folder + image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # boss health bar is always in the middle of the screen
    # and occupies 1/3 of the screen height
    height, width, c_channel = img.shape
    start_y = height // 3
    end_y = height - start_y

    start_x = width // 2
    end_x = width - start_x

    # then, get pixel color in the middle of the screen
    # to determine if the boss health bar is visible
    middle_x = width // 2
    middle_y = height // 2



    # then crop the image array to get the boss health bar
    healt_area_size = width * 0.15
    x_start = int(start_x - healt_area_size)
    end_x = int(start_x + healt_area_size)
    pixel_color = img[x_start, 70]
    t = f"Pixel color: {pixel_color}"

    # drw a rect arroun the target_pixel
    cv2.rectangle(img, (x_start, 50), (end_x, 100), (255, 0, 0), 2)
    cutted_img = img[50:100, x_start:end_x]

    ax[boss_health_images.index(image)].imshow(img)
    ax[boss_health_images.index(image)].set_title(boss_health_images.index(image))


    print(t)

fig.tight_layout()
plt.show()
```
